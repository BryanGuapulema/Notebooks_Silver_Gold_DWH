# DimVendor – Silver → Gold

**Objetivo:** Crear DimVendor en Gold desde Silver.

**Proceso:**
- Selección de VendorID → VendorKey y otras columnas importantes.
- Vista temporal tmpVendor.
- Guardar como Delta Table Gold (`Tables/DimVendor`).

```python
%%sql
CREATE OR REPLACE TEMP VIEW tmpVendor AS
SELECT
    VendorID AS VendorKey,
    VendorName,
    AccountNumber,
    CreditRating,
    ActiveFlag,
    ETL_DateInserted
FROM AdventureWorks_SilverLayer.AdventureWorks_Silver_Vendors;
```

```python
# Convertir la vista temporal a DataFrame
df = spark.sql("SELECT * FROM tmpVendor")

# Validación rápida: mostrar las primeras 10 filas
#df.show(10)

# Guardar la tabla DimVendor en Gold (Delta)
df.write.mode("overwrite").format("delta").option("overwriteSchema", "true").save("Tables/DimVendor")
```
