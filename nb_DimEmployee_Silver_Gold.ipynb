# DimEmployee – Silver → Gold

**Objetivo:** Crear la tabla DimEmployee en Gold a partir de Silver.

**Proceso:**
1. Selección de columnas relevantes, renombrando EmployeeID → EmployeeKey.
2. Crear vista temporal tmpEmployee.
3. Guardar como Delta Table Gold (`Tables/DimEmployee`).


```python
%%sql
CREATE OR REPLACE TEMP VIEW tmpEmployee AS
SELECT
    EmployeeID AS EmployeeKey,
    ManagerID,
    FirstName,
    LastName,
    FullName,
    JobTitle,
    OrganizationLevel,
    MaritalStatus,
    Gender,
    Territory,
    Country,
    `Group`,
    ETL_DateInserted
FROM AdventureWorks_SilverLayer.AdventureWorks_Silver_Employees;
```

```python
# Convertir la vista temporal a DataFrame
df = spark.sql("SELECT * FROM tmpEmployee")

# Validación rápida: mostrar las primeras 10 filas
#df.show(10)

# Guardar la tabla DimEmployee en Gold (Delta)
df.write.mode("overwrite").format("delta").option("overwriteSchema", "true").save("Tables/DimEmployee")
```
