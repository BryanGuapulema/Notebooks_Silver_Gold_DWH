# DimProductCategory – Silver → Gold

**Objetivo:** Crear DimProductCategory en Gold.

**Proceso:**
- CategoryID → CategoryKey, CategoryName, ETL_DateInserted.
- Crear vista temporal tmpProductCategory.
- Guardar como Delta Table Gold (`Tables/DimProductCategory`).


```python
%%sql
CREATE OR REPLACE TEMP VIEW tmpProductCategory AS
SELECT
    CategoryID AS CategoryKey,
    CategoryName,
    ETL_DateInserted
FROM AdventureWorks_SilverLayer.AdventureWorks_Silver_ProductCategories;
```

```python
# Convertir la vista temporal a DataFrame
df = spark.sql("SELECT * FROM tmpProductCategory")

# Validación rápida: mostrar las primeras 10 filas
#df.show(10)

# Guardar la tabla DimProductCategory en Gold (Delta)
df.write.mode("overwrite").format("delta").option("overwriteSchema", "true").save("Tables/DimProductCategory")
```
