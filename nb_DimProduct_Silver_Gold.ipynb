# DimProduct – Silver → Gold

**Objetivo:** Crear DimProduct en Gold.

**Proceso:**
- Seleccionar ProductID → ProductKey y demás columnas.
- Crear vista temporal tmpProduct.
- Guardar como Delta Table Gold (`Tables/DimProduct`).


```python
%%sql
CREATE OR REPLACE TEMP VIEW tmpProduct AS
SELECT
    ProductID AS ProductKey,
    ProductNumber,
    ProductName,
    ModelName,
    MakeFlag,
    StandardCost,
    ListPrice,
    SubCategoryID,
    ETL_DateInserted
FROM AdventureWorks_SilverLayer.AdventureWorks_Silver_Products;
```

```python
# Convertir la vista temporal a DataFrame
df = spark.sql("SELECT * FROM tmpProduct")

# Validación rápida: mostrar las primeras 10 filas
#df.show(10)

# Guardar la tabla DimProduct en Gold (Delta)

df.write.mode("overwrite").format("delta").option("overwriteSchema", "true").save("Tables/DimProduct")
```
