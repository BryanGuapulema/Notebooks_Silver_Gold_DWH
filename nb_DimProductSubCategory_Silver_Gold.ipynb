# DimProductSubCategory – Silver → Gold

**Objetivo:** Crear DimProductSubCategory en Gold.

**Proceso:**
- SubCategoryID → SubCategoryKey, CategoryID → CategoryKey, SubCategoryName.
- Crear vista temporal tmpProductSubCategory.
- Guardar como Delta Table Gold (`Tables/DimProductSubCategory`).


```python
%%sql
CREATE OR REPLACE TEMP VIEW tmpProductSubCategory AS
SELECT
    SubCategoryID AS SubCategoryKey,
    CategoryID AS CategoryKey,
    SubCategoryName
FROM AdventureWorks_SilverLayer.AdventureWorks_Silver_ProductSubcategories;
```

```python
# Convertir la vista temporal a DataFrame
df = spark.sql("SELECT * FROM tmpProductSubCategory")

# Validación rápida: mostrar las primeras 10 filas
#df.show(10)

# Guardar la tabla DimProductSubCategory en Gold (Delta)
df.write.mode("overwrite").format("delta").option("overwriteSchema", "true").save("Tables/DimProductSubCategory")
```
